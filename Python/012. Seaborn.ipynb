```python
# Importing dataset for visualisation

pkmn = pd.read_csv('Pokemon.csv')
pkmn.head()
```

### sns.distplot()

```python
# Creating a simple distribution graph

dp = sns.distplot(pkmn['Attack'],
                  bins=20,
                  color='black')

dp.set_xlabel('ATTACK')
dp.set_ylabel('DENSITY')
dp.set_title('Distribution plot')

plt.show()
```

### sns.kdeplot()

```python
# Creating a simple kde plot

sns.set_theme()

kp = sns.kdeplot(pkmn['Attack'], 
                 fill=True, 
                 color='black',
                 label='ATTACK')

kp.set_ylabel('Attack probability')
kp.set_title('Kernel density plot')
kp.legend()
plt.show()
```

### sns.rugplot()

```python
# Creating a simple rugplot

sns.set_theme(style='whitegrid')

rpa = sns.rugplot(pkmn['Attack'], 
                  height=0.1, 
                  color='r', 
                  label='Attack')

rpd = sns.rugplot(pkmn['Defense'], 
                  height=0.1, 
                  color='g', 
                  vertical=True, 
                  label='Defense')

plt.xlim(10,160)
plt.ylim(0,190)
plt.legend()
plt.show()
```

### Combining rug and kde plot

```python
# Creating a simple combination of rug and kde plot

sns.set_theme(style='darkgrid')

sns.kdeplot(pkmn['Attack'],
            pkmn['Defense'],
            color='cyan',
            shade=True)

sns.rugplot(pkmn['Attack'],
            height = 0.1,
            color = 'r')

sns.rugplot(pkmn['Defense'],
            height = 0.1,
            color = 'g',
            vertical = True)

plt.show()
```

### Advanced sns.distplot()

```python
# Combination of distribution, kde and rug

sns.set_theme(style='dark')

cp = sns.distplot(pkmn['Defense'],
                  bins=15,
                  hist = True,
                  kde = True,
                  rug = True,
                  kde_kws = {'color':'black', 
                             'ls':':', 
                              'marker':'1', 
                              'markersize':'1.5',
                              'label':'kde'},
                  hist_kws = {'color':'g', 
                              'histtype':'step',
                              'lw':3,
                              'label':'histo'},
                  rug_kws = {'color':'red',
                             'height':0.1,
                             'label':'rug'})

cp.set_xlabel('Defense points')
cp.set_ylabel('Defense frequency')
cp.set_title('Distribution')

plt.xlim(0,200)
plt.legend()
plt.show()
```

### sns.jointplot()

```python
sns.set_theme(style='ticks')

sns.jointplot(x='Attack', 
              y='Defense',
              data=pkmn,
              kind='hex', #other kinds - scatter,kde,reg,resid
              color='blue')

plt.show()
```

```python
sns.set_theme(style='darkgrid')

sns.jointplot(x ='Attack', 
              y ='Defense',
              data = pkmn,
              xlim = {0,170}, 
              kind = 'kde',     #other kinds - scatter,hex,reg,resid
              color ='black',
              fill = True)

plt.show()
```

### sns.lmplot()

```python
sns.set_theme(style='darkgrid')

lp = sns.lmplot(x='Attack',
           y = 'Defense',
           data = pkmn,
           markers='x')

plt.title('Regression plot')
plt.xlim(20,155)
```

### sns.heatmap()

```python
del pkmn['Generation']
del pkmn['Legendary']
```

##### Creating a simple heatmap

```python
sns.set_theme(style='white')

plt.subplots(figsize=(10,6))
sns.heatmap(pkmn.corr(),
            annot = True,
            fmt = '.2f',
            vmin = 0.2,
            vmax = 0.8,
            cmap = 'Purples')

plt.show()
```

##### Creating a perfect shaped heat map with border lines

```python
plt.subplots(figsize=(10,8))

sns.heatmap(pkmn.corr(),
            annot = True,
            fmt = '.2f',
            vmin = 0.2,
            vmax = 0.8,
            square =True,
            cmap = 'Blues',
            linewidth = 0.2,
            linecolor = 'black')

plt.show()
```

##### Trimming upper part of the heatmap

```python
plt.subplots(figsize=(10,8))

lmatrix = np.triu(pkmn.corr())

sns.heatmap(pkmn.corr(),
            annot = True,
            fmt ='.2f',
            vmin = 0.2,
            vmax = 0.8,
            square = True,
            linewidth = 0.2,
            cmap = 'Pastel2',
            mask = lmatrix)

plt.show()
```

##### Trimming lower part of the heatmap

```python
plt.subplots(figsize=(10,8))

umatrix = np.tril(pkmn.corr())
sns.heatmap(pkmn.corr(),
            annot = True,
            fmt = '.2f',
            vmin = 0.2,
            vmax = 0.8,
            square = True,
            linewidth = 0.2,
            cmap = 'Pastel1',
            mask = umatrix)

plt.show()
```

### Importing seaborn inbuilt dataset

```python
pn = sns.load_dataset('penguins')
pn.head()
```

```python
print(pn['species'].unique())
print(pn['island'].unique())
print(pn['sex'].unique())
```

```python
sns.pairplot(pn)
```

```python
sns.set_theme(style='darkgrid')

sns.pairplot(pn,
             x_vars=['bill_depth_mm','bill_length_mm','flipper_length_mm'],
             y_vars=['body_mass_g'],
             diag_kind=None,
             kind='scatter',
             hue='species',
             height=6,
             aspect=0.8)

plt.show()
```

```python
sns.set_theme(style='whitegrid')

sns.pairplot(pn,
             x_vars=['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm'],
             y_vars=['body_mass_g'],
             diag_kind=None,
             hue='island',
             height=5,
             aspect=0.8,
             kind='reg')

plt.show()
```

### sns.stripplot()

```python
plt.subplots(figsize=(10,8))

sns.stripplot(x='island', 
              y='body_mass_g', 
              data=pn,
              size=6,
              hue='sex',
              palette='muted',
              jitter=0.2,       # greater the jitter value higher the gap between distribution
              linewidth=1)      # adds border to the markers 

plt.show()
```

### sns.swarmplot()

```python
plt.subplots(figsize=(10,8))

sns.swarmplot(x='species',
              y='flipper_length_mm',
              data=pn,
              hue='sex',
              size=6,
              palette='husl',
              linewidth=1
             )

plt.show()
```

### sns.boxplot()

```python
plt.subplots(figsize=(15,9))

sns.set_theme()

sns.boxplot(x='species',
            y='body_mass_g',
            data= pn,
            hue='sex',
            palette = 'inferno',
            width = 0.6,
            linewidth = 1.5,
            fliersize = 10)   # size of outliers point

plt.xlabel('\nSPECIES')
plt.ylabel('BODY MASS\n')
plt.title('Statistical inference')
plt.legend()
```

```python
sns.set_theme(style='white')
sns.boxplot(data=pn,x='flipper_length_mm')
```

### sns.violinplot()

```python
plt.subplots(figsize=(15,8))

sns.violinplot(data=pn,
               x='island',
               y='flipper_length_mm',
               hue='sex',
               pallete='muted',
               split=False) # try split = True to get data in single kde

plt.xlabel('ISLAND')
plt.ylabel('FLIPPER LENGTH')
plt.title('Statistical inference')
plt.legend()

plt.show()
```
