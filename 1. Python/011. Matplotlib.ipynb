### Pyplot

### Import pyplot and creating a simple graph

```python
import matplotlib.pyplot as plt
import numpy as np

ar1 = np.arange(1,11)
ar2 = np.linspace(1000,2000,10)
plt.plot(ar1, ar2)                 # To plot the graph
plt.show()                         # To the display the output
```

### Creating graph with axis label, title, legend and grid

```python
plt.plot(ar1, ar2, label='TRAIL GRAPH')   # set legend text as label
plt.xlabel('X-Axis')                      # set x label
plt.ylabel('Y-Axis')                      # set y label
plt.title('Graph with labels and title')  # set graph title
plt.legend(loc=4)                         # display legend text ..
                                             # loc range 0-10 indicated the location of legend inside graph area
plt.grid()                     # display grids in graph 
plt.show()
```

### Creating graph with line style, maker style and colors

```python
plt.plot(ar1, ar2, color='red', linestyle=':', linewidth=3, 
         marker='o', markersize=10, markerfacecolor='black',markeredgecolor='green', label='Trail')
            # shift+tab inside plt.plot() for more line styles
            # or, run " ?? plt.plot() " for more line and marker combinations
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Graph with line style and markers')
plt.legend(loc=4)
plt.grid(color='black', linestyle='--') # shift+tab inside plt.grid() for more grid styles
plt.show()
```

### Scatter plot

```python
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x1,y1, color='blue', label='scatter points' ) # shift+tab inside plt.scatter() for help
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Graph with scatter points')
plt.legend(loc=1)
plt.grid(axis='y', color='black')
plt.show()
```

##### Creating comaprision scatter plot

```python
x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x1,y1, label='OLD DATA', marker= '2', linewidth=7, color='hotpink')
plt.scatter(x2,y2, label='NEW DATA', marker='s', color='lightgreen')

plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter ponints comparision graph')
plt.legend(loc=1)

plt.grid(axis='x')
plt.show()
```

##### Creating user defined colors for each datapoints

```python
csx = np.arange(1,11)
csy = np.random.randint(100,1000,(10,))
clr = np.array(['red','green','blue','yellow','violet','pink','gray','lavender','cyan','silver'])

plt.scatter(csx, csy, c=clr)
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points with user defined colors')

plt.show()
```

##### Creating scatterplot with colormap and colorbar

```python
msx = np.random.randint(1,100,(12,))
msy = np.random.randint(100,200,(12,))
clr = np.linspace(0,100,12)

plt.scatter(msx, msy, c=clr ,cmap='viridis')  # run "matplotlib.pyplot.colormaps()" to view list of cmaps
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points with colormap and colorbar')

plt.colorbar()                                # to display the colorbar
plt.show()
```

##### Creating different size, color scatter plot with color bar and alpha value

```python
# Alpha is used to adjust the transperency of the dots in scatter plot

ssx = np.random.randint(100, size=(15))
ssy = np.random.randint(100, size=(15))
clr = np.random.randint(100, size=(15))
sze = np.random.randint(1000, size=(15))

plt.scatter(ssx, ssy, c=clr, s=sze, alpha=0.5,cmap='nipy_spectral')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points with different size and solor')

plt.colorbar()
plt.show()
```

### Bar chart

Bar charts are used to represent the categorical datas
pyplot has the flexibility of both vertical and horizantal bar charts
plt.bar() -  vertical bar chart
plt.barh() -  horizantal bar chart

##### Vertical bar charts

```python
vbx = np.array(list('ABCDE'))
vby = np.random.randint(1,10,size=(5))

plt.bar(vbx, vby, color='tomato', width= 0.5)   # run "matplotlib.colors.cnames()" to view list of colors
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Vertical bar chart')
plt.show()
```

##### Horizantal bar charts

```python
plt.barh(vbx,vby, color='chocolate', height=0.3)  # Height to adjust the thickeness of the bar
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Horizantal bar chart')
plt.show()
```

### Objected oriented approch

##### Creating a larger graph with different line style and color

```python
fig = plt.figure(figsize= (5,4), dpi=100)  # figsize=(width, height) in inches, dpi = dots per inch
axes = fig.add_axes([0.1,0.1,0.9,0.9])     # fig.add_axes([left,bottom,height,width]) in inches
                                                # dimensions ranges between 0 to 1
                                                # left and right is distance from the margin 
                                                # height and width is plot dimension                     
        
axes.plot(ar1,ar2,'g--',label='OOA graph')                   # 'g--' represents dashed dotted lines 
                                                # shift+tab inside axes.plot() for more line styles
axes.set_xlabel('X-Axis')
axes.set_ylabel('y-Axis')
axes.set_title('Object oriented approch')
axes.legend(loc=4)
plt.show()
```

### Pie chart

##### Simple pie chart

```python
pch = np.array([35,25,15,12,13])
lbl = np.array(['Python','Java', 'Node JS', 'HTML', 'C#'])
clr = np.array(['blue', 'red', 'green', 'orange', 'purple'])

plt.pie(pch, labels=lbl, colors = clr)
plt.show()
```

##### Large exploded pie chart with legend, title and percentage

```python
fig = plt.figure(figsize=(6,6), dpi=100)
axes = fig.add_axes([0.1,0.1,0.9,0.9])

axes.pie(pch, labels=lbl, 
        colors=clr, 
        shadow=True, 
        explode = [0.2,0,0,0,0],
        autopct = '%.1f')
axes.legend(loc=2, title='Languages')
plt.title('Popular programming languages')
plt.show()
```
